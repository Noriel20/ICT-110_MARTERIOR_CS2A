##Constants in the math module: The constants in the math module include math.pi for π (pi), math.e for the mathematical constant e, math.inf for positive infinity, and math.nan for a floating-point "Not a Number" value.
## The math.exp function computes the exponential value of a number. Specifically, math.exp(x) returns e raised to the power of x, where e is the mathematical constant approximately equal to 2.71828.
##The math.pow function calculates the result of raising a number to a specified power. For instance, math.pow(x, y) returns x raised to the power of y.
#Calculating logarithms:
#Natural Logarithms (ln): Natural logarithms, or ln, can be calculated using the math.log function. For example, math.log(x) returns the natural logarithm of x.
#Logarithms base 10: To compute logarithms base 10, you can use math.log10(x). For instance, math.log10(x) returns the base-10 logarithm of x.

```python
import math


print("Value of pi:", math.pi)
print("Value of e:", math.e)
print("Positive infinity:", math.inf)
print("Not a Number (NaN):", math.nan)


x = 2
print("Exponential value of", x, ":", math.exp(x))


base = 2
exponent = 3
print(base, "raised to the power of", exponent, ":", math.pow(base, exponent))


number = 10
print("Natural logarithm of", number, ":", math.log(number))
print("Base-10 logarithm of", number, ":", math.log10(number))
```
