#### Exercise 2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E3.2

````python
hours_worked = float(input("Enter Hours: "))
hourly_rate = float(input("Enter Rate: "))

if hours_worked <= 40:
    pay = hours_worked * hourly_rate
else:
    regular_hours = 40
    overtime_hours = hours_worked - 40
    pay = (regular_hours * hourly_rate) + (1.5 * overtime_hours * hourly_rate)

print("Pay:", pay)
````

## CHALLENGE

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

    Score           Grade
    [ 97, 100]      1.0
    [ 93,  97)		1.25
    [ 89,  93)		1.5
    [ 85,  89)		1.75
    [ 81,  85)		2.0
    [ 77,  81)		2.25
    [ 74,  77)		2.5
    [ 0,   74)		3.0

Adapted from [PEW] E3.3
<br>

````python
score = float(input("Enter score between 0 and 100: "))

grade_table = {
    (97, 100): 1.0,
    (93, 97): 1.25,
    (89, 93): 1.5,
    (85, 89): 1.75,
    (81, 85): 2.0,
    (77, 81): 2.25,
    (74, 77): 2.5,
    (0, 74): 3.0
}

for key in grade_table:
    if key[0] <= score < key[1]:
        grade = grade_table[key]
        break
else:
    print("Score is out of range.")
    exit()

print("Grade:", grade)
````
