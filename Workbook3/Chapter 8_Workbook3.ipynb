#### Exercise 4

Using Index Slicing, print out the list of grades higher than 2.0

```python
grade_list=[1.0,1.25,1.5,1.75,2.0,2.25,2.5,3.0]

print(grade_list[grade_list.index(2.0) + 1:])
```

#### Exercise 5

Using Index Slicing, print out the list of grades better than 2.0.


```python
grade_list = [1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 3.0]

better_grades = [grade for grade in grade_list[:grade_list.index(1.75) + 1] if grade >= 1.0]

print(better_grades)

```

#### Exercise 6

Using Index Slicing, print out the grades 2.0,2.25,2.5.

```python
grade_list = [1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 3.0]

desired_grades = grades_list[grade_list.index(2.0):grade_list.index(2.5) + 1]

print(desired_grades)
```
