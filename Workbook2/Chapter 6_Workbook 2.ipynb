##Exercise 1


Given the code below    

    text = 'Business Analytics'

What will the following expressions produce?
    
    text[4]
    text[2:4]
    text[:4]
    text[4:]
    text[-2]
    text[-4:-2]
    text[-2:]
    text[:-2]
    

```python
#the answer

text = 'Business Analytics'

print(text[4])
print(text[2:4])
print(text[:4])
print(text[4:])
print(text[-2])
print(text[-4:-2])
print(text[-2:])
print(text[:-2])
```

#### Exercise 2

Given the code below:
    
    phrase = 'One Ring to rule them all'

What string slicing expression gives out the following words:

    'One '
    'Ring'
    'all'
    'them'
    'One Ring to rule them all'
    

```python
#answer
phrase = 'One Ring to rule them all'

p1 = phrase[:4]
p2 = phrase[4:8]
p3 = phrase[-3:]
p4 = phrase[17:21]
p5 = phrase[:]

print(p1)
print(p2)
print(p3)
print(p4)
print(p5)
```

##Challenge

Write a program which repeatedly reads a full name, then does the following:

-strips the name of any trailing spaces

-capitalizes the first letter

-puts the rest into smaller letters

```python
while True:
    name = input("Enter your full name (or type 'exit' to quit): ")

    if name.lower() == 'exit':
        print("Exiting the program.")
        break

    for_name = name.strip().capitalize()
    for_name = for_name[:1] + for_name[1:].lower()

    print('Formatted Name:', for_name)
```
